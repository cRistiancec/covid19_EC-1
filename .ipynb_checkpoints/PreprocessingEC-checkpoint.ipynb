## Convertiremos nuestros pdf a un archivo xls
Para esto deberemos registrarnos en https://pdftables.com/ de esta forma podremos hacer uso de sus servicios mediante una Api Key. Revisa un ejemplo aquí  https://pdftables.com/blog/pdf-to-excel-with-python

```python
#Obtenemos todos los paths de cada pdf
dirPDF = 'data\pdf'
pathsPDF=[]
for dirname, _, filenames in os.walk(dirPDF):
    for filename in filenames:       
        pathsPDF.append(os.path.join(dirname, filename))
```

```python
pathsPDF
```

```python
#Convertimos cada pdf a un archivo xlsx
dirXLS = 'data\\xlsx'
for path in pathsPDF:    
    fileName = path.replace(".pdf", "")
    fileName = fileName.replace("data\\pdf\\", "") 
    nameXLS = os.path.join(dirXLS, fileName)
    c = pdftables_api.Client('9f7m5kdlz9ej')
    c.xlsx(path, nameXLS) 
```
